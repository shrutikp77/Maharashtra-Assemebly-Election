```python
def categorize_polling_station(total_votes):
    if total_votes < 400:
        return 'Low Turnout'
    elif total_votes < 600:
        return 'Medium Turnout'
    elif total_votes < 800:
        return 'High Turnout'
    else:
        return 'Very High Turnout'
```

```python
df['Turnout_Category'] = df['Total Votes'].apply(categorize_polling_station)
```

```python
def categorize_competitiveness(margin, total_votes):
    margin_percent = (margin / total_votes) * 100
    if margin_percent < 5:
        return 'Highly Competitive'
    elif margin_percent < 15:
        return 'Competitive'
    elif margin_percent < 30:
        return 'Moderately Competitive'
    else:
        return 'Non-Competitive'
```

```python
df['Competitiveness'] = df.apply(lambda x: categorize_competitiveness(x['Vote_Margin'], x['Total Votes']), axis=1)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression # Import Logistic Regression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler # Often beneficial for Logistic Regression
import warnings

warnings.filterwarnings('ignore')
candidate_cols = df.columns[1:-9]
for col in candidate_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
    
candidate_votes = df[candidate_cols].values
winners = candidate_cols[np.argmax(candidate_votes, axis=1)]
df['Winner'] = winners

# --- Addressing the ValueError (for stratify) ---
# Count occurrences of each winner
winner_counts = df['Winner'].value_counts()

# Identify winners that appear only once
single_occurrence_winners = winner_counts[winner_counts == 1].index

# Filter out rows where the winner appears only once
df_filtered = df[~df['Winner'].isin(single_occurrence_winners)].copy()

print(f"Original dataset shape: {df.shape}")
print(f"Filtered dataset shape: {df_filtered.shape}")
print(f"Removed {len(single_occurrence_winners)} winners with single occurrences.")

# --- Predicting the Winner with Logistic Regression ---
print("\n=== PREDICTING WINNER (Logistic Regression) ===")

# Define features (X) and target (y) from the filtered DataFrame
X_winner = df_filtered[candidate_cols].values
y_winner = df_filtered['Winner']
scaler = StandardScaler()
X_winner_scaled = scaler.fit_transform(X_winner)
X_train_winner, X_test_winner, y_train_winner, y_test_winner = train_test_split(
    X_winner_scaled, y_winner, test_size=0.3, random_state=42, stratify=y_winner
)
lr_winner_classifier = LogisticRegression(max_iter=1000, random_state=42, solver='liblinear')
lr_winner_classifier.fit(X_train_winner, y_train_winner)

# Make predictions on the test set
y_pred_winner = lr_winner_classifier.predict(X_test_winner)

# Evaluate the model
accuracy_winner = accuracy_score(y_test_winner, y_pred_winner)
print(f"Logistic Regression Accuracy (Winner Prediction): {accuracy_winner:.2f}")

print("\nClassification Report (Winner Prediction):")
print(classification_report(y_test_winner, y_pred_winner))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
candidate_cols = df.columns[1:-9]

# Convert candidate vote columns to numeric, coercing errors to NaN and filling NaN with 0
for col in candidate_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

# Calculate total votes for each candidate across all polling stations
total_votes_per_candidate = df[candidate_cols].sum().sort_values(ascending=False)

# Get the top 5 candidates
top_5_candidates = total_votes_per_candidate.head(5)

# Calculate the total votes for all candidates (sum of all candidate votes)
total_all_candidate_votes = total_votes_per_candidate.sum()

# Calculate the vote percentage for the top 5 candidates
top_5_percentages = (top_5_candidates / total_all_candidate_votes) * 100

# Calculate the percentage for 'Others'
others_votes = total_all_candidate_votes - top_5_candidates.sum()
others_percentage = (others_votes / total_all_candidate_votes) * 100

# --- CORRECTED LINE: Use pd.concat instead of .append() ---
chart_data = pd.concat([top_5_percentages, pd.Series({'Others': others_percentage})])

# Prepare data for the pie chart
labels = chart_data.index.tolist()
sizes = chart_data.values.tolist()

# Define colors for the pie chart slices
# You can customize these colors
colors = sns.color_palette('pastel')[0:len(labels)] # Using seaborn for nice pastel colors

# --- Create the Pie Chart ---
plt.figure(figsize=(10, 9)) # Set figure size for better visualization
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140,
        pctdistance=0.85, wedgeprops=dict(width=0.4, edgecolor='w')) # Add a donut hole effect

# Draw a circle in the center to make it a donut chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Vote Percentage Distribution of Top 5 Candidates', fontsize=16, fontweight='bold')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
